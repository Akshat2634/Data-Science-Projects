#### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("seaborn")
```

#### Importing Dataset

```python
df=pd.read_csv("SampleSuperstore.csv")
```

```python
df.head()
```

#### Checking for NULL values

```python
df.isnull().sum()
```

```python
df.shape 
```

```python
#count of each category under shipment mode
df["Ship Mode"].value_counts()
```

```python
df["Country"].nunique()
```

```python
## dropping columns which dont affect profits much.
df = df.drop(["Country","Postal Code"],axis=1)
```

```python
#checking for duplicate values
df.duplicated().sum()
```

```python
#dropping the duplicate values
df.drop_duplicates(inplace=True)
df
```

#### Visualization through graphs

```python
sns.distplot(df["Quantity"])
plt.show()
```

```python
sns.distplot(df["Discount"])
plt.show()
```

```python
sns.distplot(df["Profit"])
plt.show()
```

```python
corr = df.corr()
corr
```

```python
sns.heatmap(corr, annot = True)
```

### State vs Profit

```python
sales_df = df.groupby("State")["Profit"].sum()
sales_df = sales_df.reset_index()
sales_df.head()
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x= sales_df["State"] , y = sales_df["Profit"])
plt.ylabel("Profit")
plt.xlabel("State")
plt.xticks(rotation = "vertical")
plt.show()
```

We can infer from the above barplot that the states - California and New York are having the highest profit while Ohio, Pennsylvania and Texas are having the highest losses or least profits(negative)

### Category vs Profit

```python
category_df = df.groupby("Category")["Profit"].sum()
category_df = category_df.to_frame().reset_index()
category_df
```

```python
plt.figure(figsize = (5,5))
sns.barplot(x = category_df["Category"], y = category_df["Profit"])
plt.ylabel("Profit")
plt.xlabel("Category")
plt.xticks(rotation="vertical")
plt.show()
```

We can see that the profits are least for furniture.

### Region vs Profit

```python
sns.countplot("Region", data=df)
plt.show()
```

South contributes the least to sales

### Segment vs Count

```python
sns.countplot("Segment", data=df)
plt.show()
```

Home office segment makes the least sales

### Sub-Category vs Region

```python
plt.figure(figsize=(15,7))
sns.countplot(x="Sub-Category", hue= "Region", data=df)
plt.xticks(rotation="vertical")
plt.plot()
```

Copiers, Machines and Supplies are the least sold products overall . South accounts for the least sales in any of the sub-categories.

### Sales per State

```python
df_state = df.groupby("State")["Sales"].sum().sort_values(ascending = False)
df_state = df_state.to_frame().reset_index()
df_state.head()
```

```python
df_state.plot(kind = "bar" , x = "State" ,  y = "Sales" , figsize = (12,8))
plt.ylabel("Sales")
```

1.   Highest sales- California, New York and Texas
2.   Least sales- North Dakota, West Virginia, Maine and South Dakota

### Profit and sales for sub-categories

```python
sub_df = df.groupby("Sub-Category")["Profit" , "Sales"].sum()
sub_df.plot(kind = "bar"  ,figsize = (12 , 8))
plt.show()
```
